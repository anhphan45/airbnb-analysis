## **AirBnB Listings Analysis**

#### **Data Prep**
*   Filter down the data to just listings in the city of Paris
*   Check for missing values, calculate the minimum, maximum, and average for each numeric field


```python
import pandas as pd

# The data is not formatted, and have mixed data types
data = pd.read_csv("Listings.csv", encoding="ISO-8859-1", low_memory=False, parse_dates=["host_since"])
```

```python
paris_data = data.query("city == 'Paris'").loc[:, ["host_since", "neighbourhood", "city", "accommodates", "price"]]

paris_data.info()
```

```python
paris_data.isna().sum()
```

```python
paris_data.describe()
```

```python
paris_data.query("accommodates == 0 and price == 0").count()
```

#### **Prepare for Visualzation**


```python
paris_data_neighbourhood = paris_data.groupby("neighbourhood").agg({"price": "mean"}).sort_values("price")

paris_data_neighbourhood.head()
```

```python
paris_data_accommodates = paris_data.query("neighbourhood == 'Elysee'").groupby("accommodates").agg({"price": "mean"}).sort_values("price")

paris_data_accommodates.head()
```

```python
paris_data_over_time = paris_data.set_index("host_since").resample("Y").agg({"neighbourhood": "count", "price": "mean"})

paris_data_over_time.head()
```

#### **Visualize the Data**

```python
import seaborn as sns

paris_data_neighbourhood.plot.barh(title = "Average Listing Price by Paris Neighbourhood", xlabel = "Price Per Night (Euros)", ylabel = "Neighbourhood", legend=None)

sns.despine()
```

```python
import seaborn as sns

paris_data_accommodates.plot.barh(title = "Average Listing Price by Accommodation Number", xlabel = "Price Per Night (Euros)", ylabel = "Accommodation Capacity", legend=None)

sns.despine()
```

```python
paris_data_over_time["neighbourhood"].plot(
    ylabel="New Hosts",
    title="New AirBnB Hosts in Paris Over Time"
)

sns.despine()
```

```python
paris_data_over_time["price"].plot(
    ylabel="Average Price (Euros)",
    title="Average AirBnB Price in Paris Over Time"
)

sns.despine()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(
    paris_data_over_time.index,
    paris_data_over_time["neighbourhood"],
    label="New Hosts",
    color="purple"
)

ax.set_ylabel("New Hosts")

ax2 = ax.twinx()

ax.plot(
    paris_data_over_time.index,
    paris_data_over_time["neighbourhood"],
    label="New Hosts",
    color="purple"
)

ax2.plot(
    paris_data_over_time.index,
    paris_data_over_time["price"],
    label="Average Price",
    color="pink"
)
ax2.set_ylim(0)

ax2.set_ylabel("Average Price")

ax.set_title("2015 Regulation Lead to Fewer New Hosts, Higher Prices")
```
